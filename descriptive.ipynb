```python
import pandas as pd
import numpy as np
dc3 = pd.read_csv('/content/drive/MyDrive/E_dassign/Excelr_as/assign1/Cars.csv')
dc3.head()
```

```python
m1= dc3['MPG'].mean()
s1= dc3['MPG'].std()
```

```python
prob_mpg = (1 - stats.norm.cdf(38, loc= m1, scale= s1))
print('probability of mpg > 38 is', np.round(prob_mpg, 3))
```

```python
prob_mpg2 = stats.norm.cdf(40, loc= m1, scale= s1)
print('probability of mpg < 40 is', np.round(prob_mpg2, 3))
```

```python
prob_mpg3 = (1-stats.norm.cdf(20, loc= m1, scale= s1))
print('probability of mpg > 20 is', np.round(prob_mpg3, 3))

prob_mpg4 = stats.norm.cdf(50, loc= m1, scale= s1)
print('probability of mpg < 50 is', np.round(prob_mpg4, 3))

prob_mpg5= prob_mpg4 - prob_mpg3
print('probability of mpg between 20 and 50 is', np.round(prob_mpg5, 3))
```

t scores

```python
from scipy import stats
alpha= 0.05
q= 0.95+(alpha/2)
print(np.round(q,3))
print('t score for 95% confidence interval is', np.round(stats.t.ppf(q, df= 24),3))
```

```python
alpha= 0.04
q= 0.96+(alpha/2)
print(np.round(q,3))
print('t score for 96% confidence interval is', np.round(stats.t.ppf(q, df= 24),3))
```

```python
alpha= 0.01
q= 0.99+(alpha/2)
#print(np.round(q,3))
print('t score for 99% confidence interval is', np.round(stats.t.ppf(q, df= 24),3))
```

probability

```python
import numpy as np
from scipy import stats
m= 270 # population mean
n= 18 # sample size
x= 260 # sample mean
s= 90 # sample standard deviation
ts = (x - m) / (s/ np.sqrt(n))
print('t score is', np.round(ts, 3))
t_prob= stats.t.cdf(ts, df= 17)
print('probaility is', np.round(t_prob,3)*100)
```
